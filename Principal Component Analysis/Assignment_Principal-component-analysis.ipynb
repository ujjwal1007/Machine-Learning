# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans 
```

# Load Dataset

```python
iris = pd.read_csv("Iris Dataset.csv")
```

```python
iris.head()
```

# Dataset modification acc. to given requirements

```python
species = iris['Species']
dataset = iris.drop('Species', axis=1)
dataset.head()
```

```python
# Convert dataset to numpy array

data = np.array(dataset)
```

# Covariance matrix & eigen-values/vectors

```python
# Calculate the covariance matrix
covariance_matrix = np.cov(data.T)
    
# Calculate the eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python
# Sort the eigenvectors based on the eigenvalues

sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# Project the data onto the first three eigenvectors

projected_data = np.dot(data, sorted_eigenvectors[:, :3])
```

# Plot the data-points

```python
projected_data = np.dot(data, sorted_eigenvectors[:, :3])
    
# Plot the data in the first three eigenvectors
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(projected_data[:, 0], projected_data[:, 1], projected_data[:, 2], c=species.astype('category').cat.codes, cmap='viridis')
ax.set_xlabel('1st Eigenvector')
ax.set_ylabel('2nd Eigenvector')
ax.set_zlabel('3rd Eigenvector')
ax.set
```
